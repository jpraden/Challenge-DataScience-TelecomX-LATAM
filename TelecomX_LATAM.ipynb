**2. Distribuci√≥n de la Evasi√≥n**

En este paso, el objetivo es comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes.

```python
import matplotlib.pyplot as plt
frecuencia_churn = df['Churn'].value_counts()
porcentaje_churn = df['genero'].value_counts(normalize=True) * 100

colores = ['#636efa', '#EF553B']  # Colores asignados para cada barra

fig, ax = plt.subplots(figsize=(5,3))  # Creamos la figura y los ejes con tama√±o definido

ax.set_frame_on(False)  # Ocultamos el marco del gr√°fico
ax.tick_params(axis='x', which = 'both', size = 0) # Quitamos los ticks del eje X

ax.bar(frecuencia_churn.index, frecuencia_churn.values, color=colores)  # Dibujamos las barras

ax.set_title('Distribuci√≥n de Evasi√≤n clientes (Churn)', pad=20)  # T√≠tulo con separaci√≥n ajustada
ax.set_yticks([])  # Quitamos las marcas y etiquetas del eje Y


# A√±adimos etiquetas num√©ricas encima de cada barra
for i, (total, pct) in enumerate(zip(frecuencia_churn.values, porcentaje_churn.values)):
    ax.text(i, total + 30,
            f'{total:,.0f}\n({pct:.2f}%)',          # Ubicaci√≥n y formato del valor
            color='black', fontsize=8.5,
            ha='center', va='bottom')

plt.show()
```

üîç**Resultados del an√°lisis variable "Churn" en DataFrame "df"**

- El nivel de churn (26,6%) indica que la empresa enfrenta un reto importante en retenci√≥n.
- La base estable (73,4%) es una fortaleza, pero el churn no puede ser ignorado: representa una p√©rdida sustancial de ingresos y potencial de crecimiento.

**En resumen :** Una tasa cercana al 25% suele considerarse alta en industrias de suscripci√≥n, lo que sugiere la necesidad de programas de retenci√≥n m√°s robustos.

**Clientes fugados:**
- Antig√ºedad promedio ‚âà 18 meses.

**Base general:**

- La mitad de los clientes permanece hasta 29 meses.
- El 75% alcanza hasta 55 meses de antig√ºedad.

**En resumen:** Los clientes fugados muestran una antig√ºedad significativamente menor que la base general, lo que confirma que el churn es un fen√≥meno temprano. La clave est√° en reforzar la retenci√≥n durante los primeros meses de contrato.



```python
churn_rate_by_tenure = df.groupby('tiempo_contrato')['Churn'].value_counts(normalize=True).unstack()
churn_rate_by_tenure['S√≠'] = churn_rate_by_tenure['S√≠'] * 100  # Convertir a porcentaje

fig = px.line(
    churn_rate_by_tenure.reset_index(),
    x='tiempo_contrato',
    y='S√≠',
    title='Tasa de Churn por Antig√ºedad (Tiempo de Contrato)',
    labels={'tiempo_contrato': 'Tiempo de Contrato (meses)', 'S√≠': 'Tasa de Churn (%)'}
)

fig.update_layout(xaxis_title='Tiempo de Contrato (meses)', yaxis_title='Tasa de Churn (%)')
fig.show()
```

```python
max_tenure = df['tiempo_contrato'].max()

# Define the bins for 6-month intervals. Start from 0 to cover all tenures.
bins = list(range(0, max_tenure + 7, 6))

# Create labels for the bins
labels = []
for i in range(len(bins) - 1):
    if bins[i+1] > max_tenure:
        labels.append(f'{bins[i]+1}-{max_tenure} meses')
    else:
        labels.append(f'{bins[i]+1}-{bins[i+1]} meses')

# Create the 'tiempo_contrato_intervalo' as a temporary Series, not adding it to df
tiempo_contrato_intervalo_temp = pd.cut(df['tiempo_contrato'], bins=bins, labels=labels, right=True, include_lowest=True)

# Use this temporary Series in the groupby operation
churn_rate_by_tenure_interval = df.groupby(tiempo_contrato_intervalo_temp, observed=False)['Churn'].value_counts(normalize=True).unstack()
churn_rate_by_tenure_interval['S√≠'] = churn_rate_by_tenure_interval['S√≠'] * 100 # Convert to percentage
churn_rate_by_tenure_interval = churn_rate_by_tenure_interval.reset_index()
# The reset_index() will name the column after the groupby key, which is `tiempo_contrato`.
# We need to rename it to 'tiempo_contrato_intervalo' for the plot to work with the expected label.
churn_rate_by_tenure_interval.rename(columns={'tiempo_contrato': 'tiempo_contrato_intervalo'}, inplace=True)

fig = px.line(
    churn_rate_by_tenure_interval,
    x='tiempo_contrato_intervalo',
    y='S√≠',
    title='Tasa de Churn por Intervalos de Tiempo de Contrato',
    labels={'tiempo_contrato_intervalo': 'Tiempo de Contrato (intervalos)', 'S√≠': 'Tasa de Churn (%)'},
    text='S√≠'
)

fig.update_traces(texttemplate='%{text:.2f}%', textposition='top center')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

üîç**Resultados gr√°fico "Tasa de Churn por Antig√ºedad (Tiempo de Contrato)"**

- Churn temprano: La tasa de abandono es m√°s alta en los primeros meses (1‚Äì6), con un pico muy marcado al inicio.
- Disminuci√≥n con la permanencia: A medida que aumenta el tiempo de contrato, la tasa de churn cae dr√°sticamente y se estabiliza en niveles bajos.
- Contratos mensuales: Refuerzan el riesgo de churn temprano, ya que la falta de compromiso facilita la cancelaci√≥n.
- Predictor fuerte: El tiempo de contrato muestra la correlaci√≥n negativa m√°s alta con churn (-0.35), confirmando que mayor antig√ºedad implica menor probabilidad de fuga.



________________________________________________________________________________

**En resumen:** El churn ocurre principalmente en los primeros meses de contrato; superar esa etapa cr√≠tica aumenta significativamente la permanencia y la fidelidad del cliente.

- Dispersi√≥n: Existe variabilidad en clientes de larga permanencia; no todos pagan montos altos, lo que indica diversidad de planes.
- Concentraci√≥n inicial: Muchos clientes nuevos se ubican en contratos cortos y pagos bajos, comenzando con planes b√°sicos.

**En resumen:** los clientes m√°s antiguos tienden a generar mayor ingreso mensual, lo que refuerza la importancia de la retenci√≥n a largo plazo como motor de valor y estabilidad.
